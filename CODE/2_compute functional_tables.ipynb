# Functional Table definition

**(we define the functional table as a matrix product between the GCN and the taxonomic table, and then normalite the so obtained functional table to obtain compositional values)**

### Matrix multiplication to obtain functional profiles

```python
import pandas as pd
import numpy as np
import os

# paths
inport_GCN_path              = "../DATA/GCN/remapped/GCN_eggNOG_OGs.parq"
inport_taxonomic_table_path  = "../DATA/datasets/remapped/taxonomic_table.parq"
export_functional_table_path = "../DATA/datasets/remapped/functional_table.parq"

# import
GCN = pd.read_parquet(inport_GCN_path)
taxonomic_table = pd.read_parquet(inport_taxonomic_table_path)

# matrix product
functional_table = GCN.loc[:,taxonomic_table.index] @ taxonomic_table

# drop "empty" functions
functional_table = functional_table.loc[functional_table.sum(axis=1)!=0]

# normalize to obtain compositional data
functional_table = functional_table/functional_table.sum(axis=0)

# export to parquet
functional_table.to_parquet(export_functional_table_path)

# reimport test
pd.read_parquet(export_functional_table_path).head(2)
```
